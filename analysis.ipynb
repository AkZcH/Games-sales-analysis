# Video Games Sales Analysis
### In this project, a data analysis of video games over the past years has been done, and meaningful insights are inferred. Visual statistics are also provided.

#### Step 1: Importing the necessary libraries and printing the first few rows of the data sheet after reading it 

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style to whitegrid
sns.set(style='whitegrid')


# Ensure plots are displayed inline
%matplotlib inline

# Load the dataset
df = pd.read_csv('vgsales.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Display the structure of the dataset
print("\nDataset Information:")
df.info()

# Display basic statistical metrics
print("\nStatistical Summary of the Dataset:")
print(df.describe())

# Check for missing values
print("\nMissing Values in the Dataset:")
print(df.isnull().sum())


```

### Step 2. Data Preprocessing

## **Analysis Questions**

1. What are the top-selling games globally?
2. What is the trend of global game sales over time?
3. Which regions contribute most to the sales?
4. What is the distribution of games across different platforms?


### 1. Sorting and visualising the top most selling games:

```python
sales_trend = df.groupby('Year')['Global_Sales'].sum().reset_index()

plt.figure(figsize=(14,7))
sns.lineplot(x='Year', y='Global_Sales', data=sales_trend, marker='o', color='c', hue='Year', palette='viridis', linewidth=2.5)
plt.title('Global Game Sales Trend Over Time', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Global Sales (in millions)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Year', title_fontsize='13', fontsize='11')
plt.tight_layout()
plt.show()
```

### **Insight:**
- **Wii Sports** is the top-selling game globally, with a massive lead over the other games.


### 2. Analyzing global game sales over the years 

```python
sales_trend = df.groupby('Year')['Global_Sales'].sum().reset_index()

plt.figure(figsize=(14,7))
sns.set(style='whitegrid')
plot = sns.lineplot(x='Year', y='Global_Sales', data=sales_trend, marker='o', color='royalblue', linewidth=2.5)
plt.title('Global Game Sales Trend Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Global Sales (in millions)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

### **Insight:**
- Global game sales grew significantly from the late 1980s to the early 2010s.
- There was a peak in sales around 2008-2009, likely due to the popularity of consoles like the Wii and games that appealed to a broader audience.


### 3. Regional Contribution to Sales


```python
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(10,8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), wedgeprops={'edgecolor': 'black'}, explode=[0.1, 0, 0, 0], shadow=True)
plt.title('Sales Contribution by Region', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

### **Insight:**
- **North America** contributes the most to global sales, covering almost half of the sales share, followed by **Europe**.
- **Japan** also plays a significant role, though to a lesser extent.


### 4. Distribution of Games Across Different Platforms

```python
platform_distribution = df['Platform'].value_counts().reset_index()
platform_distribution.columns = ['Platform', 'Number of Games']
platform_distribution = df['Platform'].value_counts().reset_index()
platform_distribution.columns = ['Platform', 'Number of Games']
platform_distribution = platform_distribution.sort_values(by='Number of Games', ascending=False)

plt.figure(figsize=(12,8))
sns.set(style='whitegrid')

barplot = sns.barplot(
    x='Number of Games', 
    y='Platform', 
    data=platform_distribution, 
    palette='coolwarm'
)

# Add titles and labels
plt.title('Distribution of Games by Platform', fontsize=18, weight='bold')
plt.xlabel('Number of Games', fontsize=14)
plt.ylabel('Platform', fontsize=14)

# Adding annotations to the bars
for index, value in enumerate(platform_distribution['Number of Games']):
    barplot.text(value + 1, index, f'{value}', color='black', ha="left", fontsize=12)

# Show the plot
plt.show()

```

### **Insight:**
- The dataset shows a wide variety of platforms, with certain platforms like the **PlayStation 2 (PS2)** and **Nintendo DS (DS)** having a larger number of games.
- This distribution highlights the success and popularity of these platforms over time.


## **Conclusion**

This analysis provides a comprehensive view of video game sales, uncovering key trends, regional differences, and platform distributions. The data shows that while certain games and platforms dominate the market, the gaming industry has evolved over time, with varying success across different regions.
